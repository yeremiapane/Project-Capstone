```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
from google.colab import drive

drive.mount('/content/drive')

# data harga rumah di Jakarta, Bekasi, dan Depok dari sumber rumah.com
df_jkt_rmh = pd.read_csv("/content/drive/MyDrive/data-collecting/01jakarta-rumah.csv", header=0)
df_bks_rmh = pd.read_csv("/content/drive/MyDrive/data-collecting/05bekasi-rumah.csv", header=0)
df_dpk_rmh = pd.read_csv("/content/drive/MyDrive/data-collecting/03depok-rumah.csv", header=0)
```

```python
df_jkt_rmh
```

```python
''' data rumah Jakarta, Bekasi, Depok dari sumber rumah.com'''
# menghilangkan kolom yang tidak perlu
df_jkt_rmh.drop(['nav-link','currency','hidden-sm','listing-floorarea 2','hidden-sm href 2','hidden-sm 2'],axis=1,inplace=True)
df_bks_rmh.drop(['nav-link','currency','listing-floorarea 2','hidden-sm href 2','hidden-sm 2'],axis=1,inplace=True)
df_dpk_rmh.drop(['nav-link','currency','listing-floorarea 2','hidden-sm href 2','hidden-sm 2'],axis=1,inplace=True)

# menghapus missing values
df_jkt_rmh.dropna(axis=0,inplace=True)
df_bks_rmh.dropna(axis=0,inplace=True)
df_dpk_rmh.dropna(axis=0,inplace=True)

# rename kolom
df_jkt_rmh = df_jkt_rmh.rename(columns={"listing-location": "location", "listing-floorarea": "area"})
df_bks_rmh = df_bks_rmh.rename(columns={"listing-location": "location", "listing-floorarea": "area"})
df_dpk_rmh = df_dpk_rmh.rename(columns={"listing-location": "location", "listing-floorarea": "area"})

# ubah value area menjadi data numerik
df_jkt_rmh['area'] = df_jkt_rmh.area.str.extract('(\d+)')
df_bks_rmh['area'] = df_bks_rmh.area.str.extract('(\d+)')
df_dpk_rmh['area'] = df_dpk_rmh.area.str.extract('(\d+)')

# menyamakan value location
df_jkt_rmh['location'] = "Jakarta"
df_bks_rmh['location'] = "Bekasi"
df_dpk_rmh['location'] = "Depok"

# reorder kolom
df_jkt_rmh = df_jkt_rmh.reindex(columns=['area', 'bed', 'bath', 'location', 'price'])
df_bks_rmh = df_bks_rmh.reindex(columns=['area', 'bed', 'bath', 'location', 'price'])
df_dpk_rmh = df_dpk_rmh.reindex(columns=['area', 'bed', 'bath', 'location', 'price'])
```

```python
# fungsi ubah string ke numerik
def converter(x):
  if 'jt' in x:
      return f"{(float(x.strip(' jt'))*1000000):,.2f}"
  elif 'M' in x:
      return f"{(float(x.strip(' M'))*1000000000):,.2f}"

# transformasi value price
df_jkt_rmh['p0'] = df_jkt_rmh.price.str.split("-", n=1, expand=True)[0]
df_jkt_rmh['p0'] = df_jkt_rmh['p0'].str.replace('.', '')
df_jkt_rmh['p0'] = df_jkt_rmh['p0'].str.replace(',', '.').apply(converter)

df_bks_rmh['p0'] = df_bks_rmh.price.str.split("-", n=1, expand=True)[0]
df_bks_rmh['p0'] = df_bks_rmh['p0'].str.replace('.', '')
df_bks_rmh['p0'] = df_bks_rmh['p0'].str.replace(',', '.').apply(converter)

df_dpk_rmh['p0'] = df_dpk_rmh.price.str.split("-", n=1, expand=True)[0]
df_dpk_rmh['p0'] = df_dpk_rmh['p0'].str.replace('.', '')
df_dpk_rmh['p0'] = df_dpk_rmh['p0'].str.replace(',', '.').apply(converter)


# ganti value price dengan p0
df_jkt_rmh = df_jkt_rmh.drop(columns='price')
df_jkt_rmh = df_jkt_rmh.rename(columns={"p0": "price"})

df_bks_rmh = df_bks_rmh.drop(columns='price')
df_bks_rmh = df_bks_rmh.rename(columns={"p0": "price"})

df_dpk_rmh = df_dpk_rmh.drop(columns='price')
df_dpk_rmh = df_dpk_rmh.rename(columns={"p0": "price"})


# hilangkan koma
df_jkt_rmh['price'] = df_jkt_rmh['price'].str.replace(',', '')
df_bks_rmh['price'] = df_bks_rmh['price'].str.replace(',', '')
df_dpk_rmh['price'] = df_dpk_rmh['price'].str.replace(',', '')
```

```python
# ubah tipe data
df_jkt_rmh = df_jkt_rmh.astype({"area":'float', "bed":'int', "bath": 'int', "price": 'float'}) 
df_bks_rmh = df_bks_rmh.astype({"area":'float', "bed":'int', "bath": 'int', "price": 'float'}) 
df_dpk_rmh = df_dpk_rmh.astype({"area":'float', "bed":'int', "bath": 'int', "price": 'float'}) 
```

```python
# menangani missing value pada price dengan mean lalu ganti tipe data menjadi int
df_jkt_rmh.isnull().sum()
df_jkt_rmh['price'].fillna(df_jkt_rmh['price'].mean(), inplace=True)
df_jkt_rmh['price'] = df_jkt_rmh.price.astype('int')
```

```python

df_bks_rmh.isnull().sum()
df_bks_rmh['price'].fillna(df_bks_rmh['price'].mean(), inplace=True)
df_bks_rmh['price'] = df_bks_rmh.price.astype('int')
```

```python

df_dpk_rmh.isnull().sum()
df_dpk_rmh['price'].fillna(df_dpk_rmh['price'].mean(), inplace=True)
df_dpk_rmh['price'] = df_dpk_rmh.price.astype('int')
```

```python
# gabung ketiga dataframe menjadi 1
df_rmh = pd.concat([df_jkt_rmh, df_bks_rmh, df_dpk_rmh])
```
